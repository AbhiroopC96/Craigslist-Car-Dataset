```python
#let's say we want to segregate the costs of the cars based on expensive and resonable at a price of $9000

data['isexpensive']=data.price.apply(lambda x: 'expensive' if x>9000 else 'resonable')
```

```python
#to classify the cars as new and cheap. for situations where you directly apply(like this), use axis=1

data['new_and_cheap']=data.apply(lambda x: 'yes' if x['price']<10000 and x['age']<10 else 'no', axis=1)
```
